```python
import joblib
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import PCA
import torch
import torch.nn as nn
from torch.autograd import Variable
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST
import model.dec
import matplotlib.pyplot as plt
import numpy as np

mnist_train = MNIST('./tmp/mnist', train=True, download=True, transform=transforms.ToTensor())
mnist_train_sorted = sorted(mnist_train, key=lambda x:x[1])
train_loader = DataLoader(mnist_train_sorted, batch_size=100, shuffle=False)

net = model.dec.DCEC_AutoEncoder(dim_input=28, dim_code=10, num_input_channels=1,
                 num_feature_maps0=32, num_feature_maps1=64, num_feature_maps2=128,
                 dropout_value=0, batchnorm=True)

net.load_state_dict(torch.load('./model_dropout_00_batchnorm_true_epoch_29.dump'))
```

```python
codes_array = np.zeros([60000, 10])

net.eval()
for i, data in enumerate(train_loader):
    inputs, labels = Variable(data[0]), Variable(data[1])

    outputs, codes = net(inputs)
    codes = codes.view(100, 10)
    codes_array[i*100: (i+1)*100] = codes.data.numpy()
joblib.dump(codes_array, './model_dropout_00_batchnorm_true_codes.dump')
```

```python
kmeans = MiniBatchKMeans(n_clusters=10)
clustering_result = kmeans.fit_predict(codes_array)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(codes_array)
```

```python
colors = ['black', 'gray', 'rosybrown', 'red', 'coral', 'goldenrod', 'yellow', 'green', 'blue', 'purple']
plt.figure(figsize=[12, 12])
for i in range(10):
    plt.scatter(x=pca_result[i*6000: (i+1)*6000][:, 0],
                y=pca_result[i*6000: (i+1)*6000][:, 1],
                s=1, c=colors[i], label='{}'.format(i))
plt.legend()
plt.show()
```

```python
def calculate_accuracy(clustering_result):
    number_of_correct_classification = 0
    for i in range(10):
        subset = clustering_result[i*6000: (i+1)*6000]
        assigned_index = np.max(np.bincount(subset))
        frequency_of_most_frequent_value = np.max(np.bincount(subset))
        
        print("accuracy of number {} = {}".format(i, frequency_of_most_frequent_value/6000))
        number_of_correct_classification += frequency_of_most_frequent_value
    return number_of_correct_classification / 60000
```

```python
calculate_accuracy(clustering_result)
```
