```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST
import model.dec
import matplotlib.pyplot as plt
import numpy as np

mnist_train = MNIST('./tmp/mnist', train=True, download=True, transform=transforms.ToTensor())
train_loader = DataLoader(mnist_train, batch_size=100, shuffle=True)
```

```python
def draw_figures(inputs, outputs):
    images_in = [inputs.data.numpy()[:8].flatten()[28**2 * i: 28**2 * (i+1)].reshape(28, 28) for i in range(8)]
    images_out = [outputs.data.numpy()[:8].flatten()[28**2 * i: 28**2 * (i+1)].reshape(28, 28) for i in range(8)]
    #plt.figure(figsize=[12, 12])
    plt.imshow(np.vstack([np.hstack(images_in), np.hstack(images_out)]))
    plt.title("epoch {}, loss = {}".format(epoch, loss_in_epoch))
    plt.show()
```

```python
net = model.dec.DCEC_AutoEncoder(dim_input=28, dim_code=10, num_input_channels=1,
                 num_feature_maps0=32, num_feature_maps1=64, num_feature_maps2=128,
                 dropout_value=0., batchnorm=False)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=1e-4)


loss_history = []
for epoch in range(30):
    loss_in_epoch = 0
    for i, data in enumerate(train_loader):
        inputs, labels = Variable(data[0]), Variable(data[1])
        
        outputs, codes = net(inputs)
        loss = criterion(outputs, inputs)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        loss_in_epoch += loss.data[0]
        
    draw_figures(inputs, outputs)
    torch.save(net.state_dict(), 'model_dropout_00_batchnorm_false_epoch_{:02}.dump'.format(epoch))
    loss_history.append(loss_in_epoch)
plt.plot(loss_history)
plt.show()
```
